```python
import matplotlib.pyplot as plt
import pickle as pkl
import numpy as np
import sys
```

```python
losses_idx = 0
accuracy_idx = 1

train_idx = 0
val_idx = 1
test_idx = 2
```

```python
old_metrics = None
with open('old_metrics.pkl', 'rb') as f:
    old_metrics = pkl.load(f)

old_metrics

plt.plot(range(len(old_metrics)), [train[accuracy_idx] for train, val, test in old_metrics], label="train accuracy", marker='o')
plt.plot(range(len(old_metrics)), [val[accuracy_idx] for train, val, test in old_metrics], label="validation accuracy", marker='o')
plt.plot(range(len(old_metrics)), [test[accuracy_idx] for train, val, test in old_metrics], label="test accuracy", marker='o')
plt.legend()
plt.plot()
```

```python
metrics = None
with open('new_metrics.pkl', 'rb') as f:
    metrics = pkl.load(f)

metrics

plt.plot(range(len(metrics)), [train[accuracy_idx] for train, val, test in metrics], label="train accuracy", marker='o')
plt.plot(range(len(metrics)), [val[accuracy_idx] for train, val, test in metrics], label="validation accuracy", marker='o')
plt.plot(range(len(metrics)), [test[accuracy_idx] for train, val, test in metrics], label="test accuracy", marker='o')
plt.legend()
plt.title('Accuracy curve')
plt.grid()
plt.plot()
```

```python
plt.plot(range(len(metrics)), [train[losses_idx] for train, val, test in metrics], label="train accuracy", marker='o')
plt.plot(range(len(metrics)), [val[losses_idx] for train, val, test in metrics], label="validation accuracy", marker='o')
plt.plot(range(len(metrics)), [test[losses_idx] for train, val, test in metrics], label="test accuracy", marker='o')
plt.legend()
plt.title('Loss curve')
plt.grid()
plt.plot()
```

```python
plt.plot(range(len(metrics[:12])), [val[accuracy_idx] for train, val, test in metrics[:12]], label="validation accuracy (new)", marker='o')
plt.plot(range(len(old_metrics)), [val[accuracy_idx] for train, val, test in old_metrics], label="validation accuracy (old)", marker='o')
plt.legend()
plt.grid()
plt.title('Old vs New model accuracy')
plt.show()
```
